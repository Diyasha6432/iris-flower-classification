```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('iris.csv')
x = dataset.drop(columns=['Id','Species'])
species_mapping = {'Iris-setosa': 1, 'Iris-versicolor': 2, 'Iris-virginica': 3}
inverse_species_mapping = {v: k for k, v in species_mapping.items()}
y = dataset['Species'].map(species_mapping)
dataset
```

```python
dataset.drop(columns=['Id']).hist(figsize=(10, 8))
plt.suptitle('Feature Distributions')
plt.show()
```

```python
plt.subplot(2, 2, 3)
dataset.boxplot(column=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])
plt.title('Box Plot for Each Feature')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
predicted_species = [inverse_species_mapping[pred] for pred in prediction]
print(predicted_species)
accuracy = accuracy_score(y_test, prediction)
accuracy
```
